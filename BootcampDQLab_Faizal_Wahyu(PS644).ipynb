**SUPERSTORE DATA ANALYSIS - FAIZAL WAHYU (PS644)**

The data used is the SuperStore with the following description data:

* **Order ID** Unique Order ID for each Customer.
* **Customer ID** Unique ID to identify each Customer.
* **Postal Code** Postal Code of every Customer.
* **Product ID** Unique ID of the Product.
* **Sales** Sales of the Product.
* **Quantity** Quantity of the Product.
* **Discount** Discount provided.
* **Profit** Profit/Loss incurred.
* **Category** Category of the product ordered.
* **Sub-Category** Sub-Category of the product ordered.
* **Product Name** Name of the Product.
* **Order Date** Order Date of the product.
* **Ship Date** Shipping Date of the Product.
* **Ship Mod** Shipping Mode specified by the Customer.
* **Customer Name** Name of the Customer.
* **Segment** The segment where the Customer belongs.
* **Country** Country of residence of the Customer.
* **City** City of residence of of the Customer.
* **State** State of residence of the Customer.
* **Region** Region where the Customer belong.

```python
# Import libraries and alias for easy reading
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
# Read in data in CSV format
superstore = pd.read_csv('https://raw.githubusercontent.com/faizalwahyu28/BootcampDA_DQLab/main/Superstoredataset.csv')
```

```python
# Find the number of null values for all columns
superstore.isnull().sum()
```

```python
# Find the number of duplicate data
superstore.duplicated().sum()
```

```python
# Show the duplicated rows
superstore[superstore.duplicated(keep = 'last')]
```

```python
# Drop the duplicated rows
superstore.drop_duplicates(inplace = True)
# Find the no. of rows and columns
superstore.shape
```

```python
# We will create a calculated field for Profit Margin
superstore['Profit Margin %'] = (superstore.Profit / superstore.Sales) * 100
superstore.head(5)
```

```python
# Get descriptive statistics summary
superstore.describe(include = "all")
```

**EXPLORATORY DATA ANALYSIS**

```python
# Konversi kolom 'OrderDate' ke tipe datetime
superstore['Order_Date'] = pd.to_datetime(superstore['Order_Date'])

# Ekstrak tahun dan bulan dari kolom 'OrderDate'
superstore['Year'] = superstore['Order_Date'].dt.year
superstore['Month'] = superstore['Order_Date'].dt.month

# Mengelompokkan data berdasarkan tahun dan bulan, serta menghitung total penjualan
monthly_sales = superstore.groupby(['Year', 'Month'])['Sales'].sum().reset_index()

# Mengatur ukuran gambar
plt.figure(figsize=(12, 6))

# Membuat plot tren penjualan
sns.lineplot(x='Month', y='Sales', hue='Year', data=monthly_sales)

# Menambahkan label dan judul
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Superstore Sales Trends from Year to Year')

# Menampilkan legenda
plt.legend(title='Year')

# Menampilkan grafik
plt.show()
```

**1. Which Category is Best Selling and Most Profitable?**

```python
# Group sales, profit and quantity by category
category_analysis = pd.DataFrame(superstore.groupby(['Category'])[['Sales', 'Profit', 'Quantity']].sum())
category_analysis
```

```python
# Set for grouped plots - figure with a 2x2 grid of Axes
sns.set_theme(style="whitegrid")
figure, axis = plt.subplots(1, 3, figsize=(8, 5))
# Plot barplots
cat1 = sns.barplot(x = category_analysis.index, y = category_analysis.Sales, ax=axis[0])
cat2 = sns.barplot(x = category_analysis.index, y = category_analysis.Profit, ax=axis[1])
cat3 = sns.barplot(x = category_analysis.index, y = category_analysis.Quantity, ax=axis[2])
# Set titles
cat1.set(title = 'Sales')
cat2.set(title = 'Profit')
cat3.set(title = 'Quantity')
# Rotate axis for x-axis
plt.setp(cat1.get_xticklabels(), rotation = 'vertical', size = 9)
plt.setp(cat2.get_xticklabels(), rotation = 'vertical', size = 9)
plt.setp(cat3.get_xticklabels(), rotation = 'vertical', size = 9)
# Set spacing between subplots
figure.tight_layout()
```

**2. What are the Best Selling and Most Profitable Sub-Category?**

```python
# Group by sub-category
subcat_analysis = pd.DataFrame(superstore.groupby(['Sub-Category'])[['Sales', 'Profit']].sum())
# Sort by descending order according to sales
subcat_sales = pd.DataFrame(subcat_analysis.sort_values('Sales', ascending = False))
subcat_sales
```

```python
# Sort by descending order according to profit
subcat_profit = pd.DataFrame(subcat_analysis.sort_values('Profit', ascending = False))
subcat_profit
```

```python
# Set style for plots
sns.set_theme(style="whitegrid")

# Create a figure with a 1x2 grid of Axes
figure, axis = plt.subplots(1, 2, figsize=(12, 6))

# Plot Bar Plot for Best Selling Sub-Category with a different color palette
subcat1 = sns.barplot(data=subcat_sales, x=subcat_sales.Sales, y=subcat_sales.index, ax=axis[0], palette="Set2")
subcat1.set(title="Best Selling Sub-Category")
subcat1.set_xlabel("Sales")
subcat1.set_ylabel("Sub-Category")

# Plot Bar Plot for Most Profitable Sub-Category with a different color palette
subcat2 = sns.barplot(data=subcat_profit, x=subcat_profit.Profit, y=subcat_profit.index, ax=axis[1], palette="Set2")
subcat2.set(title="Most Profitable Sub-Category")
subcat2.set_xlabel("Profit")
subcat2.set_ylabel("Sub-Category")

# Set spacing between subplots
figure.tight_layout()

# Show the plots
plt.show()
```

**3. Which is the Top Selling Sub-Category?**

```python
subcat_quantity = pd.DataFrame(superstore.groupby(['Sub-Category'])[['Quantity']].sum().sort_values('Quantity',ascending=False))
subcat_quantity
```

```python
# Plot Bar Plot for Top Selling Sub-Category
sns.set_theme(style="whitegrid")
sns.barplot(data = subcat_quantity, y = subcat_quantity.index, x = subcat_quantity.Quantity, palette = "muted")
plt.title("Top Selling Sub-Category")
plt.show()
```

**4. Which Customer Segment is Most Profitable?**

```python
segment_analysis = pd.DataFrame(superstore.groupby(['Segment'])[['Profit']].sum())
segment_analysis
```

```python
# Plot Bar Plot
sns.set_theme(style="whitegrid")
ax = sns.barplot(data=segment_analysis, x=segment_analysis.index, y=segment_analysis.Profit, palette="Set3")
plt.title("Customer Segment Profitability")

# Add value labels to each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='center', fontsize=10, color='black', xytext=(0, 5), textcoords='offset points')

plt.show()
```

```python
# Data frame segment_analysis
segment_analysis = pd.DataFrame(superstore.groupby(['Segment'])[['Profit']].sum())

# Palet warna untuk pie chart
piechart_palette = sns.color_palette("Set3", len(segment_analysis))

# Explode untuk potongan tertentu
explode = (0.1, 0, 0)

# Create a Pie Chart with customizations
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(segment_analysis['Profit'], labels=segment_analysis.index, autopct='%1.1f%%', colors=piechart_palette, wedgeprops={'edgecolor': 'none'}, explode=explode, startangle=90,shadow = True)

# Menambahkan label persen profit dan nilai profit pada setiap wedges (potongan pie)
for i, (text, autotext) in enumerate(zip(texts, autotexts)):
    autotext.set_text(f'{segment_analysis["Profit"][i]:.2f} ({autotext.get_text()})')

plt.title("Profit by Customer Segment")

plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=superstore, x='Sales', y='Profit', hue='Segment')
plt.title('Profit vs Sales by Customer Segment')
plt.show()
```

**5. Which Region is the Most Profitable?**

```python
region_analysis = pd.DataFrame(superstore.groupby(['Region'])['Profit'].sum().reset_index())
region_analysis
```

```python
# Pie Chart plot with changed color palette
explode = [0, 0, 0, 0.1]
profits = region_analysis.Profit

# Define a custom color palette
custom_colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']

# Create a Pie Chart with profit labels and percentages
plt.pie(profits, labels=region_analysis.Region, startangle=90, autopct="%1.0f%%", explode=explode, shadow=True, colors=custom_colors)
plt.title("Most Profitable by Region")
plt.show()
```

**6. Which City has the Highest Number of Sales?**

```python
city_sales = pd.DataFrame(superstore.groupby(['City'])['Sales', 'Quantity'].sum().sort_values('Sales',ascending = False))
top10 = city_sales[:10]
top10
```

```python
bottom10 = city_sales[-10:]
bottom10
```

```python
# Set for grouped plots - figure with a 1x2 grid of Axes
figure, axis = plt.subplots(1, 2, figsize=(12, 5))
sns.set_theme(style="whitegrid")

# Define custom color palettes
custom_palette_top10 = sns.color_palette("Set2", len(top10))
custom_palette_bottom10 = sns.color_palette("Pastel2", len(bottom10))

top10c = sns.barplot(data=top10, y=top10.index, x=top10.Sales, palette=custom_palette_top10, ax=axis[0])
top10c.set_title("Top 10 Cities with Highest Sales")
top10c.set_yticklabels(top10c.get_yticklabels(), size=10)

bottom10c = sns.barplot(data=bottom10, y=bottom10.index, x=bottom10.Sales, palette=custom_palette_bottom10, ax=axis[1])
bottom10c.set_title("Bottom 10 Cities with Lowest Sales")
bottom10c.set_yticklabels(bottom10c.get_yticklabels(), size=10)

# Set spacing between subplots
figure.tight_layout()
plt.show()
```

```python
# Looking for the top 5 most profitable cities
top5_profitable_cities = superstore.nlargest(5, 'Profit')

# Set for grouped plots - figure with a 1x1 grid of Axes
figure, axis = plt.subplots(1, 1, figsize=(10, 6))
sns.set_theme(style="whitegrid")

# Make a horizontal bar plot for the top 5 most profitable cities
sns.barplot(data=top5_profitable_cities, x='Profit', y='City', palette="Set2", ax=axis, orient='h')
axis.set_title("Top 5 Profitable Cities")
axis.set_xlabel("Profit")
axis.set_ylabel("City")

plt.show()
```
